```python
def bubble_sort(list1):  
    # Outer loop for traverse the entire list  
    for i in range(0,len(list1)-1):  
        for j in range(len(list1)-1):  
            if(list1[j]>list1[j+1]):  
                temp = list1[j]  
                list1[j] = list1[j+1]  
                list1[j+1] = temp  
    return list1  
  
list1 = queue. filter(['potential']) 
print("The unsorted list is: ", list1)
list1. get(0) 
print("The sorted list is: ", bubble_sort(list1))
```

```python
import moviepy.editor as mp
import librosa
import IPython.display as ipd
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from moviepy.video.io.ffmpeg_tools import ffmpeg_extract_subclip
from moviepy.editor import VideoFileClip,concatenate_videoclips
```

# Extracting the audio

```python
clip = mp.VideoFileClip("E:\DSC\PROJECTS\FIFA analysis\FIFA 21.mp4").subclip(1, 1380)
clip.audio.write_audiofile("E:\\DSC\\PROJECTS\\FIFA analysis\\audio.wav")
filename = "E:\\DSC\\PROJECTS\\FIFA analysis\\audio.wav"
```

```python
# loading the file with a sampling rate
x, sr = librosa.load(filename, sr=16000)
# To get duration of the audio clip in minutes
int(librosa.get_duration(x, sr) / 60)
# Dividing into chunks of 5 seconds
max_slice = 10
window_length = max_slice * sr
# Playing the audio chunk
a = x[21 * window_length:22 * window_length]
ipd.Audio(a, rate=sr)
```

# Computing Short-time energy

```python
s_energy = np.array([sum(abs(x[i:i + window_length] ** 2)) for i in range(0, len(x), window_length)])
plt.hist(s_energy)
plt.show()
```

# Classifying every chunk based on threshold value

```python
df = pd.DataFrame(columns=['energy', 'start', 'end'])
thresh = 200
row_index = 0
for i in range(len(s_energy)):
    value = s_energy[i]
    if value >= thresh:
        i = np.where(s_energy == value)[0]
        df.loc[row_index, 'energy'] = value
        df.loc[row_index, 'start'] = i[0] * 10
        df.loc[row_index, 'end'] = (i[0] + 1) * 10
        row_index = row_index + 1
temp = []
i = 0
j = 0
n = len(df) - 2
m = len(df) - 1
while (i <= n):
    j = i + 1
    while (j <= m):
        if (df['end'][i] == df['start'][j]):
            df.loc[i, 'end'] = df.loc[j, 'end']
            temp.append(j)
            j = j + 1
        else:
            i = j
            break
df.drop(temp, axis=0, inplace=True)
print(df)
        
       
```

# Merging all the video clips and creating video highlights

```python
 start = np.array(df['start'])
end = np.array(df['end'])
for i in range(len(df)):
    if i != 0:
        start_lim = start[i] - 10
    else:
        start_lim = start[i]
        end_lim = end[i]
        filename = "highlight" + str(i + 1) + ".mp4"
        ffmpeg_extract_subclip("E:\DSC\PROJECTS\FIFA analysis\FIFA 21.mp4", start_lim, end_lim, targetname=filename)
```
